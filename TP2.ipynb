```python
# Import libriraies
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student-mat.csv")
```

```python
df.head()
```

First we will analyse the population of student (répartition of age, sexe...) 

```python

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
counts = df['sex'].value_counts()
counts.plot.pie(autopct='%.1f%%')
plt.show()
```

In the Data set there is a little bit more data of women then man. But we can consider that there are equal. So we can compare the two sexe for the next analyse.

```python
fig, axes = plt.subplots(1, 2, figsize= (14,8))
plt.suptitle('Alcohol consumption per gender')

sns.boxplot(x="sex",y="Walc", data=df,ax=axes[0])

sns.boxplot(x="sex",y="Dalc", data=df,ax=axes[1])
```

We can see that men drinks more that women in general.So the question is: Does women's result on exam are better then men's result?
we can also see that student in général drinks more alcohol on weekend but only a few of them are drinking during working day. It seems normal. 

```python
#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
counts = df['age'].value_counts()
counts.plot.pie(autopct='%.1f%%')
plt.show()
```

The majority of student in the data set are minor or they just turned 18 yo. So they are very young and not allowed to drink alcohol.

```python
df.corr()

f,ax=plt.subplots(figsize=(18,18))
sns.heatmap(df.corr(),annot=True,linewidth=0.5,fmt='.2f',ax=ax)
plt.show()
```

As we can see from the correlation map that there is a hight correlation between the variable goout and Walc. So student that often go out with their friends probably also drink a lot on Weekend. Some of them also drink on working day with a smallest correslation. There is a high correlation between Walc and Dalc. So many student who drink on weekend also drink during working day. 
We can also see that exam score are highly correlated with other exam score. It says that students have almost the same grade at each exams. So we can compare for the next analyse the result with just one exam. Nevertheless we want to know if there is some parameters that infer with success. We can see that the Mother and Father study and the study time affects the most the success of the student. The negative correlation beetween Medu and Fedu with failure confirm these analyse. It's seem logical, indeed we know that our social environment affect our success. And it's logical that student who study a lot have more chance to success. We can also notice negative correlation beetween exam score and failures. So we can interpret this result by, less you have failures and more your exam score are better. We can see the same correlation for studytime and Walc and Dalc. So more you are drinking and less you have time to study.
There is also a negative correlation with exam score and go out and Walc Dalc. So student which go out and drink alcool on weekend or on work day probably have bad result. But the most curious thing we can see is that the negative correlation is bigger for goout than drink on weekend. And same for drinking on weekend than drinking on work day. So the principal impact on success is the goout factor. We have seen above that goout have correlation with Walc and Dalc. So indirectly we can say that alcohol have an impact on success. 


```python
fig, axes = plt.subplots(1, 2, figsize= (14,8))
plt.suptitle('Alcohol consumption per age and grade at G1')

sns.violinplot(ax=axes[0], x='Walc', y='G1', data=df, color='cyan')
axes[0].set_title('Daily alcohol consumption')

sns.violinplot(ax=axes[1], x='Dalc', y='G1', data=df, color='blue')
axes[1].set_title('Weekend alcohol consumption');
```

Buy these graph we cannot say that alcohol consumption have bad effect on the result. We can see that people who drink a little (for Walc=2) have better chance to have good exams score. So maybe drink a little bit on weekend make people more relax and more efficient for the other things. 

```python
plt.figure(figsize=(16,5))
plt.title(" G1 results depending on absences")
sns.scatterplot(y="G1", x="absences", data = df , palette = 'summer')
```

As we can see the higher score are majority on the left so student which has hight exams score are rarely absent.
